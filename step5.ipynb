```python
url = "https://fantasy.premierleague.com/api/fixtures/?future=1"
response = requests.get(url)
future_fixtures = pd.DataFrame(response.json())
future_fixtures.to_csv("data/future_fixtures.csv", index=False)
print("Future fixtures (GW30-GW38) saved!")
```
